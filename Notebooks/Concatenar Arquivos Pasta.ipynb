```python
from os import listdir
from os.path import isfile, join
import pandas as pd
```

```python
caminho = "datasets\isolamento\ES"
saida = "datasets\isolamento"

arqs = [ join(caminho, f) for f in listdir(caminho) if isfile(join(caminho, f))]
#print(arqs)
ls = []
for arq in arqs:
    df = pd.read_csv(arq, index_col=None, header=0, encoding="utf-8")
    ls.append(df)
    
dadosES = pd.concat(ls, ignore_index=True, axis=0)   
dadosES
```

```python
dadosES.to_csv(join(saida, "ES.csv"), index=False)
```
